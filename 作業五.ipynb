```python
import numpy as np
import matplotlib.pyplot as plt

# 定義方程式
def equation(x, a, b, c):
    return a * x**2 + b * x + c

# 定義數值範圍
x = np.linspace(1, 150, 150)  # 生成1到150之間的150個均勻分佈的數值

# 第一組條件
a = -2
b = 4
c = 5

# 計算方程式對應的y值
y = equation(x, a, b, c)

# 繪製圖形
plt.plot(x, y, label='Equation')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Equation Graph')
plt.grid(True)

plt.axhline(0, color='black', linewidth=0.5)  # 繪製 x 軸

# 最高點和最低點
max_index = np.argmax(y)
min_index = np.argmin(y)
max_x = x[max_index]
max_y = y[max_index]
min_x = x[min_index]
min_y = y[min_index]
plt.plot(max_x, max_y, 'ro', label=f'Highest Point ({round(max_x)}, {round(max_y)})')
plt.plot(min_x, min_y, 'go', label=f'Lowest Point ({round(min_x)}, {round(min_y)})')
plt.text(max_x, max_y, f'({round(max_x)}, {round(max_y)})', verticalalignment='bottom')
plt.text(min_x, min_y, f'({round(min_x)}, {round(min_y)})', verticalalignment='top')

zero_indices = np.where(np.isclose(y, 0, atol=1e-6))[0]
zero_x = x[zero_indices]
zero_y = y[zero_indices]
plt.plot(zero_x, zero_y, 'bo', label='Intersection with x-axis')

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 第二組條件
a2, b2, c2 = 3, 2, -6

# x值範圍
x = np.arange(1, 501)

# 計算第二組條件下的y值
y2 = a2 * x ** 2 + b2 * x + c2

# 找出最高點和最低點的索引
max_index = np.argmax(y2)
min_index = np.argmin(y2)

# 最高點的座標
max_point = (x[max_index], y2[max_index])
# 最低點的座標
min_point = (x[min_index], y2[min_index])

# 繪製xy軸圖
plt.plot(x, y2, label='Group 2: a=3, b=2, c=-6')

# 標註最高點和最低點的座標值
plt.scatter(max_point[0], max_point[1], color='red')
plt.scatter(min_point[0], min_point[1], color='blue')
plt.annotate(f'Max: {max_point}', xy=max_point, xytext=(max_point[0], max_point[1] - 10), color='red')
plt.annotate(f'Min: {min_point}', xy=min_point, xytext=(min_point[0], min_point[1] + 10), color='blue')

# 加上十字XY軸
plt.axhline(0, color='black')
plt.axvline(0, color='black')

# 設定十字XY軸的樣式
plt.axhline(0, color='black', linestyle='-', linewidth=0.5)
plt.axvline(0, color='black', linestyle='-', linewidth=0.5)

# 加上網格
plt.grid(True, linestyle='--', linewidth=0.5)

# 加上標籤和圖例
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

# 顯示圖形
plt.show()
```
